```python
# Check the distribution of gene symbols
# Create a copy of the series with NaN values replaced by "Missing Gene Symbol"
gene_symbol_with_missing = df['Gene Symbol'].fillna("Missing Gene Symbol")
gene_symbol_counts = gene_symbol_with_missing.value_counts()

# Display basic statistics
print(f"Total number of rows: {len(df)}")
print(f"Number of unique gene symbols: {df['Gene Symbol'].nunique()}")
print(f"Number of rows with missing gene symbols: {df['Gene Symbol'].isna().sum()}")

# Display the top 20 most frequent gene symbols (including missing)
print("\nTop 20 most frequent gene symbols (including missing):")
print(gene_symbol_counts.head(20))

# Plot the distribution of the top 30 gene symbols
plt.figure(figsize=(12, 8))
gene_symbol_counts.head(30).plot(kind='bar')
plt.title('Top 30 Most Frequent Gene Symbols (Including Missing)')
plt.xlabel('Gene Symbol')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Check if there are any patterns in the distribution
print("\nDistribution statistics:")
print(f"Mean occurrences per gene: {gene_symbol_counts.mean():.2f}")
print(f"Median occurrences per gene: {gene_symbol_counts.median():.2f}")
print(f"Max occurrences: {gene_symbol_counts.max()} (Gene: {gene_symbol_counts.idxmax()})")
print(f"Min occurrences: {gene_symbol_counts.min()} (Gene: {gene_symbol_counts.idxmin()})")
```

```python
# First, let's make sure we're only looking at rows with valid gene symbols
df_with_genes = df[~df['Gene Symbol'].isna()]

# Group by Gene Symbol and calculate mean scores for increased proliferation
gene_proliferation_scores = df_with_genes.groupby('Gene Symbol')['Score - increased proliferation (automatic)'].mean().reset_index()

# Sort to find highest and lowest scoring genes
highest_proliferation = gene_proliferation_scores.sort_values('Score - increased proliferation (automatic)', ascending=False)
lowest_proliferation = gene_proliferation_scores.sort_values('Score - increased proliferation (automatic)', ascending=True)

# Display top 10 genes with highest proliferation scores
print("Top 10 genes with HIGHEST proliferation scores:")
print(highest_proliferation.head(10))

# Display top 10 genes with lowest proliferation scores
print("\nTop 10 genes with LOWEST proliferation scores:")
print(lowest_proliferation.head(10))

# Create a visualization of the top and bottom genes
plt.figure(figsize=(14, 8))

# Plot top 10 highest scoring genes
plt.subplot(1, 2, 1)
sns.barplot(x='Score - increased proliferation (automatic)', y='Gene Symbol', 
            data=highest_proliferation.head(10), palette='Reds_r')
plt.title('Top 10 Genes with Highest Proliferation Scores')
plt.xlabel('Mean Proliferation Score')
plt.tight_layout()

# Plot top 10 lowest scoring genes
plt.subplot(1, 2, 2)
sns.barplot(x='Score - increased proliferation (automatic)', y='Gene Symbol', 
            data=lowest_proliferation.head(10), palette='Blues_r')
plt.title('Top 10 Genes with Lowest Proliferation Scores')
plt.xlabel('Mean Proliferation Score')
plt.tight_layout()

plt.show()

# Let's also check the distribution of scores for our top 3 most common genes
common_genes = ['KIF11', 'INCENP', 'COPB1']
for gene in common_genes:
    gene_data = df[df['Gene Symbol'] == gene]
    mean_score = gene_data['Score - increased proliferation (automatic)'].mean()
    
    print(f"\nProliferation score statistics for {gene}:")
    print(f"Mean score: {mean_score:.4f}")
    print(f"Min score: {gene_data['Score - increased proliferation (automatic)'].min():.4f}")
    print(f"Max score: {gene_data['Score - increased proliferation (automatic)'].max():.4f}")
    
    plt.figure(figsize=(8, 4))
    sns.histplot(gene_data['Score - increased proliferation (automatic)'].dropna(), kde=True)
    plt.axvline(mean_score, color='red', linestyle='--', label=f'Mean: {mean_score:.4f}')
    plt.title(f'Distribution of Proliferation Scores for {gene}')
    plt.xlabel('Score - increased proliferation (automatic)')
    plt.legend()
    plt.show()
```
